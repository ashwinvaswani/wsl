```python
import pandas as pd
```

```python
df = pd.read_csv('original.csv')
df.head()
```

```python
df['Id'] = df.apply(lambda row: row['Image Index'].split('.')[0], axis=1)
df['all'] = df.apply(lambda row: row['Finding Labels'].split('|'), axis=1)
conditions = list(set(sum(df['all'].to_list(), [])))
for condition in conditions:
    df[condition] = df.apply(lambda row: 1 if condition in row['all'] else 0, axis=1)
df['Male_Female'] = df.apply(lambda row: 0 if row['Patient Gender'] == 'M' else 1, axis=1)
df['AP_PA'] = df.apply(lambda row: 0 if row['View Position'] == 'AP' else 1, axis=1)
df.head()
```

```python
df['all'] = df.apply(lambda row: {'Male_Female':int(row['Male_Female']),
                                  'AP_PA':int(row['AP_PA']),
                                  'No_Finding':int(row['No Finding']),
                                  'Hernia':int(row['Hernia']),
                                  'Emphysema':int(row['Emphysema']),
                                  'Cardiomegaly':int(row['Cardiomegaly']),
                                  'Nodule':int(row['Nodule']),
                                  'Mass':int(row['Mass']),
                                  'Edema':int(row['Edema']),
                                  'Consolidation':int(row['Consolidation']),
                                  'Pneumonia':int(row['Pneumonia']),
                                  'Atelectasis':int(row['Atelectasis']),
                                  'Pneumothorax':int(row['Pneumothorax']),
                                  'Effusion':int(row['Effusion']),
                                  'Pleural_Thickening':int(row['Pleural_Thickening']),
                                  'Fibrosis':int(row['Fibrosis']),
                                  'Infiltration':int(row['Infiltration'])}, axis=1)

# re-ordering columns
df = df[['Id', 'Patient ID', 'OriginalImage[Width', 'Height]', 'all', 'Patient Age', 'Male_Female', 'AP_PA', 'No Finding', 'Hernia',
         'Emphysema', 'Cardiomegaly', 'Nodule', 'Mass', 'Edema', 'Consolidation', 'Pneumonia',
         'Atelectasis', 'Pneumothorax', 'Effusion', 'Pleural_Thickening', 'Fibrosis', 'Infiltration']]

# re-naming columns
df.columns = ['Id', 'Patient', 'width', 'height', 'all', 'Age', 'Male_Female', 'AP_PA', 'No_Finding', 'Hernia',
         'Emphysema', 'Cardiomegaly', 'Nodule', 'Mass', 'Edema', 'Consolidation', 'Pneumonia',
         'Atelectasis', 'Pneumothorax', 'Effusion', 'Pleural_Thickening', 'Fibrosis', 'Infiltration']

df.to_csv('info.csv', index=False)
df.head()
```

```python
df_bbox = pd.read_csv('original_bbox.csv')
df_bbox['Id'] = df_bbox.apply(lambda row: row['Image Index'].split('.')[0], axis=1)
df_bbox = pd.merge(df_bbox, df, on='Id', how='inner')
df_bbox['x2'] = df_bbox.apply(lambda row: row['Bbox [x'] + row['w'], axis=1)
df_bbox['y2'] = df_bbox.apply(lambda row: row['y'] + row['h]'], axis=1)
df_bbox = df_bbox[['Id', 'width', 'height', 'Bbox [x', 'y', 'x2', 'y2', 'Finding Label']]
df_bbox.columns = ['Id', 'width', 'height', 'x1', 'y1', 'x2', 'y2', 'Label']
df_bbox.to_csv('info_bbox.csv', index=False)
df_bbox.head()
```

```python
df_tv = pd.read_csv('train_val_list.txt', header=None)
tv = df_tv.apply(lambda row: row[0].split('.')[0], axis=1).to_list()
tv = list(set(df[df['Id'].isin(tv)].Patient.to_list()))
df_t = pd.read_csv('test_list.txt', header=None)
t = df_t.apply(lambda row: row[0].split('.')[0], axis=1).to_list()
t = list(set(df[df['Id'].isin(t)].Patient.to_list()))
len(tv), len(t)
```
