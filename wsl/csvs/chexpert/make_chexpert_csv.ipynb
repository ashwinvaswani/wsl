```python
import pandas as pd
```

```python
df = pd.read_csv('original.csv')
df.head()
```

```python
df['Path'] = df.apply(lambda row: row['Path'][len('CheXpert-v1.0-small/train/'):-len('.jpg')], axis=1)
df['Patient'] = df.apply(lambda row: row['Path'].split('/')[0], axis=1)
df.head()
```

```python
df['Sex'] = df.apply(lambda row: 0 if row['Sex'] == 'Male' else 1, axis=1)
df['Frontal/Lateral'] = df.apply(lambda row: 0 if row['Frontal/Lateral'] == 'Frontal' else 1, axis=1)
df['AP/PA'] = df.apply(lambda row: 0 if row['AP/PA'] == 'AP' else 1, axis=1)
df = df.fillna(0)
df = df.replace(-1, 1)
df.head()
```

```python
df['all'] = df.apply(lambda row: {'Male_Female':int(row['Sex']),
                                  'Frontal_Lateral':int(row['Frontal/Lateral']),
                                  'AP_PA':int(row['AP/PA']),
                                  'No_Finding':int(row['No Finding']),
                                  'Enlarged_Cardiomediastinum':int(row['Enlarged Cardiomediastinum']),
                                  'Cardiomegaly':int(row['Cardiomegaly']),
                                  'Lung_Opacity':int(row['Lung Opacity']),
                                  'Lung_Lesion':int(row['Lung Lesion']),
                                  'Edema':int(row['Edema']),
                                  'Consolidation':int(row['Consolidation']),
                                  'Pneumonia':int(row['Pneumonia']),
                                  'Atelectasis':int(row['Atelectasis']),
                                  'Pneumothorax':int(row['Pneumothorax']),
                                  'Pleural_Effusion':int(row['Pleural Effusion']),
                                  'Pleural_Other':int(row['Pleural Other']),
                                  'Fracture':int(row['Fracture']),
                                  'Support_Devices':int(row['Support Devices'])}, axis=1)
df.head()
```

```python
# re-ordering columns
df = df[['Patient', 'Path', 'all', 'Age', 'Sex', 'Frontal/Lateral', 'AP/PA', 'No Finding',
       'Enlarged Cardiomediastinum', 'Cardiomegaly', 'Lung Opacity',
       'Lung Lesion', 'Edema', 'Consolidation', 'Pneumonia', 'Atelectasis',
       'Pneumothorax', 'Pleural Effusion', 'Pleural Other', 'Fracture',
       'Support Devices']]
# re-naming columns
df.columns = ['Patient', 'Id', 'all', 'Age', 'Male_Female', 'Frontal_Lateral', 'AP_PA', 'No_Finding',
       'Enlarged_Cardiomediastinum', 'Cardiomegaly', 'Lung_Opacity',
       'Lung_Lesion', 'Edema', 'Consolidation', 'Pneumonia', 'Atelectasis',
       'Pneumothorax', 'Pleural_Effusion', 'Pleural_Other', 'Fracture',
       'Support_Devices']
df.astype({'Age':int,
           'Male_Female':int,
           'Frontal_Lateral':int,
           'AP_PA':int,
           'No_Finding':int,
           'Enlarged_Cardiomediastinum':int,
           'Cardiomegaly':int,
           'Lung_Opacity':int,
           'Lung_Lesion':int,
           'Edema':int,
           'Consolidation':int,
           'Pneumonia':int,
           'Atelectasis':int, 
           'Pneumothorax':int,
           'Pleural_Effusion':int,
           'Pleural_Other':int,
           'Fracture':int,
           'Support_Devices':int}, inplace=True)
df.to_csv('info.csv', index=False)
```
